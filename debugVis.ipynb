```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
with open('./debug/loss_data.txt', 'r') as f:
    lines = f.readlines()
series = lines[0].strip().split(",")
datas = np.array([list(map(float, line.strip().split(","))) for line in lines[1:]])
x = [i for i in range(len(datas))]

num_losses = len(series)
fig, axes = plt.subplots(int(num_losses / 3) + 1, 3, figsize=(10 * 3, 10 * int(num_losses / 3) + 1))  # 
if num_losses == 1:  # lossaxeslist
    axes = [axes]

for i in range(int(num_losses / 3) + 1):
    for j in range(3):
        if i * 3 + j < num_losses:
            axes[i][j].plot(datas[:, i * 3 + j], label=series[i * 3 + j])
            axes[i][j].set_title(series[i * 3 + j])
            axes[i][j].grid(True)
            axes[i][j].set_ylim(bottom=0)

plt.tight_layout()
plt.show()
```
